```python
from PIL import Image
import numpy as np

from IPython.display import display
```

```python
image_path = '/content/drive/MyDrive/big_image.jpg'
```

```python
image = Image.open(image_path)

print("Formato da Imagem: ", image.format)
print("Modo da Imagem: ", image.mode)
print("Dimensões da Imagem: ", image.size)

display(image)
```

```python
image_array = np.array(image)

height,width, _ = image_array.shape

gray_array = np.zeros((height,width), dtype =np.uint8)

for i in range(height):
  for j in range(width):

    r,g,b = image_array[i,j]

    gray_value = int(0.3 * r + 0.59 * g + 0.11 * b)

    gray_array[i, j] = gray_value


gray_image = Image.fromarray(gray_array, mode= 'L')

display(gray_image)
```

```python
gray_image = image.convert('L')
display(gray_image)
```

```python
gray_array = np.array(gray_image)

threshold = 110

bw_array = np.zeros_like(gray_array, dtype=np.uint8)

bw_array[gray_array < threshold] = 0
bw_array[gray_array >= threshold] = 255

bw_image = Image.fromarray(bw_array, mode ='L')

display(bw_image)
```

```python
gray_image.save("/content/drive/MyDrive/imagem_cinza.png")

bw_image.save("/content/drive/MyDrive/imagem_pb.png")
```

#### Comparando o tamanho das imagens

```python
import os

size_color = os.path.getsize("/content/drive/MyDrive/big_image_color.jpg")
size_mb_color = size_color / (1024*1024)

size_gray = os.path.getsize("/content/drive/MyDrive/big_image_gray.png")
size_mb_gray = size_gray / (1024*1024)

size_bw = os.path.getsize("/content/drive/MyDrive/big_image_bw.png")
size_mb_bw = size_bw / (1024*1024)

print("Tamanho da imagem colorida: {:.2f} MB".format(size_mb_color))
print("Tamanho da imagem em escala de cinza: {:.2f} MB".format(size_mb_gray))
print("Tamanho da imagem em preto e branco: {:.2f} MB".format(size_mb_bw))
```
